## Observação sobre o modelo
No MongoDB, documentos da mesma coleção não precisam ter exatamente os mesmos campos. Essa flexibilidade permite que produtos de diferentes categorias sejam armazenados juntos sem a obrigação de um esquema fixo.

## Parte 1: Exercícios Práticos

### Exercício 1: Criação e Coleções (O "C" do CRUD)
1. Conecte ao MongoDB com `mongosh`.
2. Crie um banco de dados chamado `loja_virtual`:
```js
use loja_virtual
```
3. Crie a coleção `produtos` e insira um produto simples usando `insertOne()`.

````python
db.produtos.insertOne({
  nome: "Smartphone Galaxy A15",
  categoria: "Eletronicos",
  preco: 1299.90,
  marca: "Samsung",
  armazenamento: "128GB",
  cor: "Azul"
})
````

4. Insira pelo menos dois outros produtos de categorias diferentes usando `insertMany()`.

````python
db.produtos.insertMany([
  {
    nome: "MongoDB na Pratica",
    categoria: "Livros",
    preco: 79.90,
    autor: "Joao Silva",
    editora: "Tech Books",
    paginas: 320
  },
  {
    nome: "Camiseta Basica",
    categoria: "Roupas",
    preco: 49.90,
    tamanho: "M",
    cor: "Preta",
    material: "Algodao"
  }
])
````

### Exercício 2: Consultas e Filtros (O "R" do CRUD)
- Liste todos os produtos da coleção.
- Busque apenas produtos onde o preço seja superior a 100.
- Liste apenas os produtos da categoria `Eletronicos`.
- Realize uma busca que retorne apenas o `nome` e o `preco` dos produtos.

````python
db.produtos.find().pretty()
````

````python
db.produtos.find({ preco: { $gt: 100 } }).pretty()
````

````python
db.produtos.find({ categoria: "Eletronicos" }).pretty()
````

````python
db.produtos.find({}, { nome: 1, preco: 1, _id: 0 }).pretty()
````

### Exercício 3: Atualização Dinâmica (O "U" do CRUD)
- Atualize o preço de um produto específico usando o operador `$set`.
- Adicione um novo campo chamado `estoque` para todos os produtos usando `updateMany()`.
- Marque os produtos da categoria `Roupas` com o campo `promocao: true`.

````python
db.produtos.updateOne(
  { nome: "Smartphone Galaxy A15" },
  { $set: { preco: 1199.90 } }
)
````

````python
db.produtos.updateMany({}, { $set: { estoque: 50 } })
````

````python
db.produtos.updateMany(
  { categoria: "Roupas" },
  { $set: { promocao: true } }
)
````

### Exercício 4: Exclusão de Dados (O "D" do CRUD)
- Remova um produto específico usando `deleteOne()`.
- Remova todos os produtos de uma categoria usando `deleteMany()`.

````python
db.produtos.deleteOne({ nome: "Tenis Runner Pro" })
````

````python
db.produtos.deleteMany({ categoria: "Livros" })
````
